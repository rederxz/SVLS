```python
import matplotlib.pyplot as plt
```

```python
ET = [0.8125, 0.8197, 0.8120, 0.8100, 0.8194]
TC = [0.8581, 0.8506, 0.8668, 0.8498, 0.8547]
WT = [0.8946, 0.8916, 0.8962, 0.8957, 0.8971]
SVLS_ET = [0.8127] * 5
SVLS_TC = [0.8625] * 5
SVLS_WT = [0.8926] * 5
rates = ['0.67 : 0.33', '0.75 : 0.25', '0.83 : 0.17', '0.88 : 0.12', '0.90 : 0.10']
x = range(len(rates))

plt.figure(figsize=(8, 6))
plt.plot(ET, label='ET')
plt.plot(TC, label='TC')
plt.plot(WT, label='WT')
plt.xticks(x, rates, rotation=0)
plt.ylim(0.80, 0.90)
plt.plot(SVLS_ET, label='ET (0.5:0.5)', linestyle="--" )
plt.plot(SVLS_TC, label='TC (0.5:0.5)', linestyle="--" )
plt.plot(SVLS_WT, label='WT (0.5:0.5)', linestyle="--" )
plt.legend(loc=(12/16, 12.7/20))
plt.xlabel('center weight : round weight')
plt.ylabel('DSC')
plt.savefig('dsc_diff_rates.svg', dpi=600)
plt.show()
```
