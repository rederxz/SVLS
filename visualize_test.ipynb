```python
import matplotlib.pyplot as plt
```

```python
ET = [0.8125, 0.8197, 0.8120, 0.8100, 0.8194]
TC = [0.8581, 0.8506, 0.8668, 0.8498, 0.8547]
WT = [0.8946, 0.8916, 0.8962, 0.8957, 0.8971]
SVLS_ET = [0.8127] * 5
SVLS_TC = [0.8625] * 5
SVLS_WT = [0.8926] * 5

plt.plot(ET, label='ET')
plt.plot(TC, label='TC')
plt.plot(WT, label='WT')
plt.plot(SVLS_ET, label='SVLS_ET')
plt.plot(SVLS_TC, label='SVLS_TC')
plt.plot(SVLS_WT, label='SVLS_WT')
plt.legend()

plt.show()
```
